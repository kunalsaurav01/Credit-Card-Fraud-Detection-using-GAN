Imports

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

Load Dataset

```python
data = pd.read_csv("creditcard.csv")
data.head()
```

Separate Fraud & Normal Transactions

```python
fraud_data = data[data['Class'] == 1]
normal_data = data[data['Class'] == 0]

print("Fraud samples:", fraud_data.shape)
print("Normal samples:", normal_data.shape)
```

Feature Scaling

```python
scaler = StandardScaler()

fraud_features = fraud_data.drop("Class", axis=1)
fraud_scaled = scaler.fit_transform(fraud_features)

input_dim = fraud_scaled.shape[1]
print("Input Dimension:", input_dim)
```

Generator Model

```python
def build_generator(input_dim):
    model = models.Sequential([
        layers.Dense(256, input_shape=(input_dim,)),
        layers.LeakyReLU(0.2),
        layers.BatchNormalization(),

        layers.Dense(128),
        layers.LeakyReLU(0.2),
        layers.BatchNormalization(),

        layers.Dense(input_dim, activation="tanh")
    ])
    return model
```

Discriminator Model

```python
def build_discriminator(input_dim):
    model = models.Sequential([
        layers.Dense(256, input_shape=(input_dim,)),
        layers.LeakyReLU(0.2),
        layers.Dropout(0.3),

        layers.Dense(128),
        layers.LeakyReLU(0.2),
        layers.Dropout(0.3),

        layers.Dense(1, activation="sigmoid")
    ])
    return model
```

GAN Class (Custom Training Loop)

```python
class GAN(tf.keras.Model):
    def __init__(self, generator, discriminator):
        super(GAN, self).__init__()
        self.generator = generator
        self.discriminator = discriminator

    def compile(self, g_optimizer, d_optimizer, loss_fn):
        super(GAN, self).compile()
        self.g_optimizer = g_optimizer
        self.d_optimizer = d_optimizer
        self.loss_fn = loss_fn

    @tf.function
    def train_step(self, real_data):
        batch_size = tf.shape(real_data)[0]
        noise = tf.random.normal((batch_size, input_dim))

        # Label smoothing
        real_labels = tf.random.uniform((batch_size, 1), 0.8, 1.0)
        fake_labels = tf.random.uniform((batch_size, 1), 0.0, 0.2)

        # --------------------
        # Train Discriminator
        # --------------------
        with tf.GradientTape() as d_tape:
            fake_data = self.generator(noise, training=True)

            real_output = self.discriminator(real_data, training=True)
            fake_output = self.discriminator(fake_data, training=True)

            d_real_loss = self.loss_fn(real_labels, real_output)
            d_fake_loss = self.loss_fn(fake_labels, fake_output)

            d_loss = d_real_loss + d_fake_loss

        grads = d_tape.gradient(d_loss, self.discriminator.trainable_variables)
        self.d_optimizer.apply_gradients(zip(grads, self.discriminator.trainable_variables))

        # ----------------
        # Train Generator
        # ----------------
        noise = tf.random.normal((batch_size, input_dim))
        valid_labels = tf.ones((batch_size, 1))

        with tf.GradientTape() as g_tape:
            generated_data = self.generator(noise, training=True)
            fake_output = self.discriminator(generated_data, training=True)
            g_loss = self.loss_fn(valid_labels, fake_output)

        grads = g_tape.gradient(g_loss, self.generator.trainable_variables)
        self.g_optimizer.apply_gradients(zip(grads, self.generator.trainable_variables))

        return {"d_loss": d_loss, "g_loss": g_loss}
```

Compile GAN

```python
generator = build_generator(input_dim)
discriminator = build_discriminator(input_dim)

gan = GAN(generator, discriminator)

gan.compile(
    g_optimizer=optimizers.Adam(0.0001, beta_1=0.5),
    d_optimizer=optimizers.Adam(0.0002, beta_1=0.5),
    loss_fn=tf.keras.losses.BinaryCrossentropy()
)
```

Prepare Dataset

```python
fraud_dataset = tf.data.Dataset.from_tensor_slices(fraud_scaled.astype("float32"))
fraud_dataset = fraud_dataset.shuffle(1024).batch(128)
```

Train GAN

```python
epochs = 10000
d_losses = []
g_losses = []

for epoch in range(epochs):
    for batch in fraud_dataset:
        losses = gan.train_step(batch)

    d_losses.append(losses["d_loss"].numpy())
    g_losses.append(losses["g_loss"].numpy())

    if epoch % 1000 == 0:
        print(f"Epoch {epoch} | D Loss: {d_losses[-1]:.4f} | G Loss: {g_losses[-1]:.4f}")

    if np.mean(g_losses[-50:]) > 10:
        print("Early stopping due to unstable generator")
        break
```

Generate Synthetic Fraud Samples

```python
num_samples = 5000
noise = tf.random.normal((num_samples, input_dim))
synthetic_fraud = generator(noise, training=False).numpy()

synthetic_fraud = scaler.inverse_transform(synthetic_fraud)
```

Create Final Balanced Dataset

```python
synthetic_df = pd.DataFrame(synthetic_fraud, columns=fraud_features.columns)
synthetic_df["Class"] = 1

normal_df = normal_data.copy()

final_data = pd.concat([normal_df, synthetic_df])
final_data = final_data.sample(frac=1).reset_index(drop=True)

print(final_data["Class"].value_counts())
```

Plot Losses

```python
plt.figure(figsize=(10,5))
plt.plot(d_losses, label="Discriminator Loss")
plt.plot(g_losses, label="Generator Loss")
plt.legend()
plt.title("GAN Training Losses")
plt.show()
```
